```python
x = data_1.drop('FVC',axis = 1).values[:,:]
y = data_1['FVC'].values[:]

x_train,x_test,y_train,y_test = model_selection.train_test_split(x,y)
```

```python
model = linear_model.LinearRegression()
model.fit(x_train,y_train)

model.score(x,y),model.score(x_train,y_train),model.score(x_test,y_test)
```

```python
model = linear_model.RidgeCV(alphas = [0.1,1,10,100],cv = 4)
model.fit(x,y)
model.alpha_,model.best_score_
```

```python
model = ensemble.RandomForestRegressor(n_estimators = 300,ccp_alpha = 10)
model.fit(x_train,y_train)

model.score(x,y),model.score(x_train,y_train),model.score(x_test,y_test)
```

```python
model = ensemble.GradientBoostingRegressor(n_estimators = 300,max_depth = 5)
model.fit(x_train,y_train)

model.score(x,y),model.score(x_train,y_train),model.score(x_test,y_test)
```

```python
model = ensemble.GradientBoostingRegressor(n_estimators = 500,max_depth = 4,learning_rate = 0.05)
model.fit(x_train,y_train)

model.score(x,y),model.score(x_train,y_train),model.score(x_test,y_test)
```

```python
model = xgb.XGBRegressor(n_estimators = 200,max_depth = 4)
model.fit(x_train,y_train)

model.score(x,y),model.score(x_train,y_train),model.score(x_test,y_test)
```

```python
estimators = [
    ('GB',ensemble.GradientBoostingRegressor(n_estimators = 300,max_depth = 5)),
    ('XGB',xgb.XGBRegressor(n_estimators = 200,max_depth = 4)),
    ('RF',ensemble.RandomForestRegressor(n_estimators = 300,ccp_alpha = 10))
]
model = ensemble.StackingRegressor(estimators = estimators,final_estimator = linear_model.LinearRegression())
cv = model_selection.cross_validate(model,x,y,cv = 4)
cv
```
