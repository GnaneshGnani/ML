```python
x = train_data.drop('label',axis = 1).values[:,:].reshape(-1,28,28,1).astype(np.float64) / 255
y = train_data['label'].values[:]
test_x = test_data.values[:,:].reshape(-1,28,28,1).astype(np.float64)/ 255

x_train,x_test,y_train,y_test = model_selection.train_test_split(x,y,train_size = 0.9)
x_train.shape,y_train.shape,x_test.shape,test_x.shape
```

```python
model = keras.models.Sequential()
model.add(keras.layers.Conv2D(filters = 6,kernel_size = (5,5),strides = (1,1),activation = 'relu'))
model.add(keras.layers.AvgPool2D(pool_size = (2,2),strides = (2,2)))
model.add(keras.layers.Conv2D(filters = 6,kernel_size = 5,strides = 1,activation = 'relu'))
model.add(keras.layers.AvgPool2D(pool_size = 2,strides = 2))
model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(120,kernel_initializer = 'he_uniform',activation = 'relu'))
model.add(keras.layers.Dense(84,kernel_initializer = 'he_uniform',activation = 'relu'))
model.add(keras.layers.Dense(10,activation = 'softmax'))
model.compile(optimizer = keras.optimizers.Adam(),metrics = ['accuracy'],loss = keras.losses.SparseCategoricalCrossentropy())
history = model.fit(x_train,y_train,epochs = 50,batch_size = 8,validation_data = (x_test,y_test))
```

```python
plt.grid()
plt.plot(range(50),history.history['accuracy'])
plt.plot(range(50),history.history['val_accuracy'],color = 'r')
```

```python
plt.plot(range(50),history.history['loss'])
```

```python
y_pred = model.predict(test_x)
y_pred = np.argmax(y_pred,axis = 1)
y_pred
submission = pd.DataFrame({'ImageId' : range(1,test_data.shape[0] + 1),})
submission['Label'] = y_pred
submission
submission.to_csv('CNN Submission.csv',index = False)
```
