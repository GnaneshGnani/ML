```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import linear_model, svm, preprocessing, metrics, feature_selection, model_selection, tree
```

```python
data = pd.read_csv('train.csv')
data.head()
```

```python
data.info()
```

```python
data = data.drop(columns = ['Id'])
```

```python
features = []
string_features = []
for i in data.columns:
    if data[i].values.dtype == object:
        string_features.append(i)
    else:
        features.append(i) 
print(data[features].values[:,:], data[features].values.shape)
```

```python
selection_1 = feature_selection.VarianceThreshold(threshold = 0.5)
selection_1.fit(data[features].values[:,:])
data_1 = selection_1.transform(data[features].values[:,:])
print(data_1.shape)
x = 0
best = set()
for i in data_1[0]:
    while i != data[features].values[0,x]:
        x += 1
    best.add(x)
    print(x, end = " ")
print()
x = 0
for i in data_1[1]:
    while i != data[features].values[1,x]:
        x += 1
    print(x, end = " ")
selected_features = [data[features].columns[i] for i in best]
selected_features
```

```python
plt.figure(figsize = (30,30))
sns.heatmap(data[features].corr(), annot = True, cmap = 'Blues')
```

```python
plt.figure(figsize = (30,20))
sns.heatmap(data[selected_features].corr(), annot = True, cmap = "Blues")
```

```python
important_features = ['OverallQual', 'GrLivArea', 'GarageCars', 'GarageArea']
extra_features = ['YearBuilt', 'TotalBsmtSF']
multi_correlation = [('GarageYrBlt', 'YearBuilt'), ('TotRmsAbvGrd', 'GrLivAere'), ('TotRmsAbvGrd', 'BedroomAbvGr'), 
                     ('TotalBsmtSF', '1stFlrSF')]
```

```python
non_linear_data = data.drop(columns = important_features + extra_features)
non_linear_data.columns
non_linear_features = [i for i in non_linear_data.columns if i not in string_features]
data[non_linear_features]
plt.figure(figsize = (10,250))
for i in range(len(non_linear_features)):
    plt.subplot(len(non_linear_features), 1, i + 1)
    sns.scatterplot(x = non_linear_features[i], y = 'SalePrice', data = data)
```

```python
notes = ['ScreenPorch', 'Fireplaces', 'FullBath', '1stFlrSF', 'BsmtFinSF1']
plt.figure(figsize = (5, 7))
plt.subplot(3, 1, 1)
sns.heatmap(data[notes + ['SalePrice']].corr(), cmap = 'Blues', annot = True)
plt.figure(figsize = (7, 13))
plt.subplot(3, 1, 2)
sns.heatmap(data[notes + important_features + extra_features + ['SalePrice']].corr(), cmap = 'Blues', annot = True)
plt.figure(figsize = (5, 7))
plt.subplot(3, 1, 3)
final_features = ['TotalBsmtSF', 'OverallQual', 'YearBuilt', 'GarageCars', 'GrLivArea']
sns.heatmap(data[final_features + ['SalePrice']].corr(), annot = True, cmap = 'Blues')
```

```python
new_data = data
new_data.dropna()
temp_list = list(new_data.columns[new_data.isna().sum() > 0])
new_data = new_data.drop(temp_list, axis = 1)
new_data = pd.get_dummies(new_data, dtype = int)
new_data
x = new_data.drop(columns = ['SalePrice']).values[:,:]
y = new_data['SalePrice'].values[:]
print(x.shape, y.shape)
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y, random_state = 0)
model = linear_model.LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print(model.score(x, y), model.score(x_train, y_train), model.score(x_test, y_test))
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
